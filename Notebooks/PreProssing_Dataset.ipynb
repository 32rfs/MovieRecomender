### The file has the Data for 2017 as we already had the years priors in the first Dataset, for the following years we need to get in other datasets.

```python
df_new_meta = df_movies_meta.loc[df_movies_meta.year <= 2017,['genres','id','title','year']]
```

```python
df_new_meta.head(10)
```

```python
df_new_meta['id'] = df_new_meta['id'].astype(int)
```

```python
df_data = pd.merge(df_new_meta, df_credits, on='id')
df_data
```

```python
df_data['genres'] = df_data['genres'].map(lambda x: ast.literal_eval(x))
df_data['crew'] = df_data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))

def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
df_data['genres_list'] = df_data['genres'].map(lambda x: make_genresList(x))
df_data['director_name'] = df_data['crew'].map(lambda x: get_directors(x))
```

```python
df_data['director_name'].head(10)
```

```python
df_data['genres_list'].head(10)
```

```python
df_movie = df_data.loc[:,['director_name','genres_list','title']]
```

```python
df_movie
```

```python
df_movie['comb'] = df_movie['director_name'] +' ' + df_movie['genres']
```
